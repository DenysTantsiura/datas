```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, 
# які вказали мову програмування Python
python_data = pd.read_csv('2017_jun_final.csv')
python_data = python_data[python_data['Язык.программирования'] == 'Python']
# python_data.drop([idx for idx in python_data.index if python_data.iloc[idx]
#                  ['Язык.программирования'] != 'Python'], axis=0, inplace=True)
print(python_data)
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
print(python_data.shape)
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком 'Посада'
print(python_data_gby := python_data.groupby(['Должность']))
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком 'Посада', виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику 'Зарплата.в.місяць'
# Зарплата.в.месяц == salary  # має ж бути однакове, дублювання?
print(python_data_salary := python_data_gby.agg(
    {'Зарплата.в.месяц': ['min', 'max']}))
```

```python
labels = np.array(python_data_salary.index)

data = np.array(python_data_salary['Зарплата.в.месяц']['max'])
plt.pie(
    data,
    labels=labels,
    shadow=False,
    autopct="%.2f%%",
    pctdistance=0.8,
    labeldistance=1.1,
)

plt.title('Відносна межа найбільших зарплат (Python) за посадами.', fontsize=12.5)
plt.show()
```

```python
labels = np.array(python_data_salary.index)

data = np.array(python_data_salary['Зарплата.в.месяц']['min'])
explode = [0.15 for _ in python_data_salary['Зарплата.в.месяц']['min']]
plt.pie(
    data,
    labels=labels,
    shadow=False,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=0.80,
    labeldistance=1.0,
)

plt.title('Відносна межа найменших зарплат (Python) за посадами.', fontsize=12.5)
plt.show()
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик 'avg'
# де, у початковій таблиці (для Python) чи таблиці зарплат за посадою?
def fill_avg_salary(df: Series) -> float | None:
    return df.describe().loc['mean']

# у початковій таблиці (для Python)
python_data['avg'] = np.ones(python_data.shape[0]) * pd.DataFrame(
    python_data['Зарплата.в.месяц']).apply(fill_avg_salary, axis=0).loc['Зарплата.в.месяц']
print(python_data.head())
print(python_data['avg'])
```

```python
# таблиці зарплат за посадою
python_data_salary['avg'] = python_data_salary.apply(fill_avg_salary, axis=1)
print(python_data_salary)
```

```python
# у початковій таблиці (для Python) з врахуванням посади, а не загальну середню серед усіх
# 
def fill_avg_salary_by_position(df: Series) -> float | None:
    return df.groupby(['Должность'])['Зарплата.в.месяц'].mean()


python_data['avg_by_position'] = np.ones(python_data.shape[0]) * pd.DataFrame(
    python_data['Зарплата.в.месяц']).apply(fill_avg_salary, axis=0).loc['Зарплата.в.месяц']
print(python_data.head())
print(python_data['avg_by_position'])
```

```python
plt.bar(
    python_data_salary.index,
    python_data_salary['avg'],
)

plt.xlabel('Посади', fontsize='small', color='midnightblue')
plt.ylabel('Середня зарплата, $', fontsize='small', color='midnightblue')
plt.title('Середня зарплата для Python відносно меж, за посадами', fontsize=12)
plt.xticks(rotation=80)
plt.grid(True)
plt.subplots_adjust(
    top=0.935,
    bottom=0.49,
    left=0.14,
    right=0.81,
    hspace=0.2,
    wspace=0.2
)
plt.show()
```
