```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = pd.read_csv('2017_jun_final.csv')
python_data.drop([idx for idx in python_data.index if python_data.iloc[idx]
                 ['Язык.программирования'] is np.nan], axis=0, inplace=True)
print(python_data)
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
print(python_data.shape)
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком 'Посада'
print(python_data_gby := python_data.groupby(['Должность']))
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком 'Посада', виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику 'Зарплата.в.місяць'
print(python_data_salary := python_data_gby.agg({'salary': ['min', 'max']}))
```

```python
labels = np.array(python_data_salary.index)

data = np.array(python_data_salary['salary']['max'])
plt.pie(
    data,
    labels=labels,
    shadow=False,
    autopct="%.2f%%",
    pctdistance=1.10,
    labeldistance=1.25,
)

plt.title('Відносна межа найбільших зарплат за посадами.', fontsize=12.5)
plt.show()
```

```python
labels = np.array(python_data_salary.index)

data = np.array(python_data_salary['salary']['min'])
explode = [0.15 for _ in python_data_salary['salary']['min']]
plt.pie(
    data,
    labels=labels,
    shadow=False,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=0.850,
    labeldistance=1.0,
)

plt.title('Відносна межа найменших зарплат за посадами.', fontsize=12.5)
plt.show()
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик 'avg'
# Чи таки не просто середнє за місяць, а середнє за посадою? чи середнє за мовою? 
# Новий стовпчик до початкової таблиці? Чи до попереднього результату?
def fill_avg_salary(df: Series) -> float | None:
    return df.describe().loc['mean']


python_data['avg'] = np.ones(python_data.shape[0]) * pd.DataFrame(
    python_data['salary']).apply(fill_avg_salary, axis=0).loc['salary']
print(python_data.head())
print(python_data['avg'])
```

```python
python_data_salary['avg'] = python_data_salary.apply(fill_avg_salary, axis=1)
print(python_data_salary)
```

```python
plt.bar(
    python_data_salary.index,
    python_data_salary['avg'],
)

plt.xlabel('Посади', fontsize='small', color='midnightblue')
plt.ylabel('Середня зарплата, $', fontsize='small', color='midnightblue')
plt.title('Середня зарплата відносно меж, за посадами', fontsize=12)
plt.xticks(rotation=80)
plt.grid(True)
plt.subplots_adjust(
    top=0.935,
    bottom=0.49,
    left=0.14,
    right=0.81,
    hspace=0.2,
    wspace=0.2
)
plt.show()
```

```python
def fill_avg_salary_by_position(df: DataFrame, el: str) -> float | None:
    df_c = df[df.loc[el]['Должность'] == df['Должность']]
    return df_c['salary'].agg(salary='mean')['salary']


python_data['avg_by_Должность'] = [fill_avg_salary_by_position(
    python_data, el) for el in python_data.index]
print(python_data['avg_by_Должность'])
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
print(python_data['avg'].describe())
print(python_data_salary['avg'].describe())
print(python_data['avg_by_Должность'].describe())
```
